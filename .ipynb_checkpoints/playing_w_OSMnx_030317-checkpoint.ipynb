```python
import matplotlib.pylab as plt
import osmnx as ox
import numpy as np

np.random.seed(222)

%matplotlib inline
%pylab inline
```

```python
G = ox.graph_from_point((40.69, -73.98), distance=1000, network_type='drive')
ox.plot_graph(G)
```

```python
intersections = [key for key, value in G.graph['streets_per_node'].items() if value>1]
nc = ['r' if node in intersections else 'none' for node in G.nodes()]
ox.plot_graph(G, node_color=nc)
```

```python
basic_stats = ox.basic_stats(G)
print basic_stats
```
